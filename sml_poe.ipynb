```r
data_symmetric <- rnorm(1000, mean = 0, sd = 1) # Normal distribution (symmetric)
data_right_skewed <- rlnorm(1000, meanlog = 0, sdlog = 1) # Log-normal (Right Skew)
data_left_skewed <- -rlnorm(1000, meanlog = 0, sdlog = 1) # Negated log-normal (Left Skew)
data_mesokurtic <- rnorm(1000, mean = 0, sd = 1) # Normal distribution (Kurtosis = 3)
data_leptokurtic <- rt(1000, df = 3) # t-distribution with df=3
data_platykurtic <- runif(1000, min = -2, max = 2) # Uniform Distribution 
```

```r
dataset <- list(
    "Symmetric" = data_symmetric,
    "Right-Skewed" = data_right_skewed,
    "Left-Skewed" = data_left_skewed,
    "Mesokurtic" = data_mesokurtic,
    "Leptokurtic" = data_leptokurtic,
    "Platykurtic" = data_platykurtic
)
```

```r
par(mfrow = c(3, 2))

for (name in names(dataset)) {
    data <- dataset[[name]]
    sk <- round(skewness(data), 2)
    ku <- round(kurtosis(data), 2)

    # create histogram
    hist(data, main = paste(name, "\nSkewness:", sk, "Kurtosis:", ku), probability = TRUE,
         xlab = "Value", col="salmon", border = "white", breaks = 30)
    # add density curve
    lines(density(data), col = "lightblue", lwd = 2)
}

par(mfrow = c(1, 1))
```

### Exp 07: Implement Conditional Probability and Bayes Theorem

```r
# Install necessary packages if not already installed
if (!require(ggplot2)) {
  install.packages("ggplot2")
  library(ggplot2)
}

# Bayes' Theorem - Disease Test
set.seed(123)
P_Disease <- 0.01 # 1% of the population has the disease
P_Positive_given_Disease <- 0.95 # Sensitivity (True Positive Rate)
P_Positive_given_NoDisease <- 0.05 # False Positive Rate

# Total probability of testing positive (whether disease or not)
P_Positive <- (P_Positive_given_Disease * P_Disease) +
  (P_Positive_given_NoDisease * (1 - P_Disease))

# Probability of having the disease given a positive test (Bayes' Theorem)
P_Disease_given_Positive <- (P_Positive_given_Disease * P_Disease) / P_Positive

cat("Probability of having the disease given a positive test:", round(P_Disease_given_Positive, 4), "\n\n")

# Data for True Positive and False Positive Probabilities
df_bayes <- data.frame(
  Category = c("True Positive", "False Positive"), 
  Probability = c(P_Positive_given_Disease * P_Disease, P_Positive_given_NoDisease * (1 - P_Disease))
)

# Visualization of Bayes' Theorem (True vs False Positives)
ggplot(df_bayes, aes(x = Category, y = Probability, fill = Category)) +
  geom_bar(stat = "identity", color = "black") +
  ggtitle("Bayes' Theorem: True vs False Positives") +
  ylab("Probability") +
  theme_minimal()

# Conditional Probability Example - Dice Rolls
P_Even <- 3/6 # Probability of rolling an even number {2,4,6}
P_Four_given_Even <- 1/3 # Probability of rolling a 4 given the number is even
P_Four_and_Even <- P_Even * P_Four_given_Even
P_Four_if_Even <- P_Four_and_Even / P_Even

cat("Probability of rolling a 4 given that the number is even:", round(P_Four_if_Even, 4), "\n\n")

# Data for Dice Roll Probabilities
df_dice <- data.frame(
  Outcome = factor(c(1, 2, 3, 4, 5, 6)),
  Probability = rep(1/6, 6)  # Each outcome has an equal probability
)

# Visualization of Dice Roll Probabilities
ggplot(df_dice, aes(x = Outcome, y = Probability, fill = Outcome)) +
  geom_bar(stat = "identity", color = "black") +
  ggtitle("Dice Roll Probabilities") +
  ylab("Probability") +
  theme_minimal()
```

### Exp 08: Estimating Geometric Probability within a unit square and circle 

```r
# Install ggplot2 if not already installed
if (!require(ggplot2)) {
  install.packages("ggplot2")
  library(ggplot2)
}

# Define the geometric probability simulation function
geometric_probability <- function(total_points, radius = 0.5) {
  # Generate random points in the unit square
  x <- runif(total_points, 0, 1)
  y <- runif(total_points, 0, 1)

  # Calculate distances from the center (0.5, 0.5)
  distances <- sqrt((x - 0.5)^2 + (y - 0.5)^2)

  # Count points within the circle
  points_in_circle <- sum(distances <= radius)

  # Calculate the probability
  probability <- points_in_circle / total_points

  # Create a data frame for plotting
  df <- data.frame(x = x, y = y, inside_circle = distances <= radius)

  # Create the plot
  plot <- ggplot(df, aes(x, y, color = inside_circle)) +
    geom_point(alpha = 0.6) +
    scale_color_manual(values = c("red", "blue"), labels = c("Outside", "Inside")) +
    theme_minimal() +
    labs(
      title = "Geometric Probability Simulation",
      subtitle = paste("Radius =", radius, "| Points =", total_points),
      x = "X Coordinate",
      y = "Y Coordinate",
      color = "Point Location"
    ) +
    theme(legend.position = "top")

  # Print the plot
  print(plot)

  # Return the estimated probability
  return(probability)
}

# Run the simulation
set.seed(123) # For reproducibility
total_points <- 10000
radius <- 0.5
prob <- geometric_probability(total_points, radius)

# Print the estimated probability
cat("Estimated probability:", prob, "\n")
```

### Exp 11: Fit a Linear regression model using ordinary least square

```r
# Load necessary libraries
library(ggplot2)
library(dplyr)
library(patchwork)

# Use mtcars dataset
df <- mtcars
x <- df$wt
y <- df$mpg

# Fit linear model
lm_fit <- lm(y ~ x, data = df)

# Predictions with confidence and prediction intervals
pred_df <- data.frame(x = seq(min(x), max(x), length.out = 100))
pred <- predict(lm_fit, newdata = pred_df, interval = "confidence")
pred_pi <- predict(lm_fit, newdata = pred_df, interval = "prediction")

# Prepare data frame for plotting
pred_df <- pred_df %>%
  mutate(fit = pred[, "fit"],
         ci.lwr = pred[, "lwr"],
         ci.upr = pred[, "upr"],
         pi.lwr = pred_pi[, "lwr"],
         pi.upr = pred_pi[, "upr"])

# Model statistics
metrics <- list(
  Equation = paste0("y = ", round(coef(lm_fit)[1], 2), " + ", round(coef(lm_fit)[2], 2), " * x"),
  R2 = paste0("R² = ", round(summary(lm_fit)$r.squared, 3)),
  PValue = paste0("p = ", format.pval(summary(lm_fit)$coefficients[2, 4], digits = 2))
)

# Main plot
main_plot <- ggplot(df, aes(x = x, y = y)) +
  geom_ribbon(data = pred_df, aes(x = x, ymin = pi.lwr, ymax = pi.upr),
              fill = "gray90", alpha = 0.5, inherit.aes = FALSE) +
  geom_ribbon(data = pred_df, aes(x = x, ymin = ci.lwr, ymax = ci.upr),
              fill = "gray70", alpha = 0.5, inherit.aes = FALSE) +
  geom_point(color = "#1f77b4", size = 3, alpha = 0.7) +
  geom_line(data = pred_df, aes(x = x, y = fit),
            color = "#ff7f0e", size = 1.5, inherit.aes = FALSE) +
  geom_smooth(method = "loess", se = FALSE,
              color = "#2ca02c", linetype = "dashed", size = 1) +
  labs(title = "mpg vs. wt",
       x = "wt", y = "mpg") +
  theme_minimal(base_size = 12) +
  theme(plot.title = element_text(face = "bold", size = 14)) +
  annotate("text", x = min(df$wt) + 0.05 * diff(range(df$wt)),
           y = max(df$mpg), label = paste(metrics$Equation, metrics$R2, metrics$PValue, sep = "\n"),
           hjust = 0, vjust = 1, size = 4)

# Residual plot
res_plot <- df %>%
  mutate(predicted = predict(lm_fit),
         residuals = residuals(lm_fit)) %>%
  ggplot(aes(x = predicted, y = residuals)) +
  geom_point(color = "#1f77b4", size = 3, alpha = 0.7) +
  geom_hline(yintercept = 0, color = "red", linetype = "dashed") +
  labs(title = "Residuals vs Predicted", x = "Predicted", y = "Residuals") +
  theme_minimal(base_size = 12)

# Show both plots side by side
main_plot + res_plot
```

### Exp 12: Multiple regression 

```r
# Load necessary libraries
library(MASS)    # For Boston Housing dataset
library(car)     # For VIF calculation
library(caret)   # For train-test split

# Load the Boston dataset
data(Boston)
head(Boston)

# Separate features (X) and target (y)
X <- Boston[, -14]  # All columns except 'medv' (target)
y <- Boston$medv     # Median home value

# Split into training and testing datasets (80% training, 20% testing)
set.seed(42)  # For reproducibility
train_index <- createDataPartition(y, p = 0.8, list = FALSE)
X_train <- X[train_index, ]
X_test <- X[-train_index, ]
y_train <- y[train_index]
y_test <- y[-train_index]

# Create training dataset
train_data <- data.frame(X_train, medv = y_train)

# Fit the linear regression model
model <- lm(medv ~ ., data = train_data)
summary(model)

# Calculate VIF to check for multicollinearity
vif_values <- vif(model)
vif_df <- data.frame(Feature = names(vif_values), VIF = vif_values)
print(vif_df)

# Drop 'rad' and 'tax' (high VIF)
X_train_reduced <- X_train[, !(colnames(X_train) %in% c("rad", "tax"))]
X_test_reduced <- X_test[, !(colnames(X_test) %in% c("rad", "tax"))]

# Create reduced training dataset
train_data_reduced <- data.frame(X_train_reduced, medv = y_train)

# Refit the model with reduced features
model_reduced <- lm(medv ~ ., data = train_data_reduced)

# Check VIF again after feature reduction
vif_reduced <- vif(model_reduced)
print(data.frame(Feature = names(vif_reduced), VIF = vif_reduced))

# Predict on test data
test_data_reduced <- data.frame(X_test_reduced)
y_pred <- predict(model_reduced, newdata = test_data_reduced)

# Calculate R^2 and RMSE
r2 <- cor(y_test, y_pred)^2
rmse <- sqrt(mean((y_test - y_pred)^2))

# Output the results
cat("R2:", r2, "\nRMSE:", rmse)
```

### Exp 13: Implement Unconstrained Optimization problem using gradient descent

```r
# Define the quadratic function and its gradient
f <- function(x) (x - 3)^2
grad_f <- function(x) 2 * (x - 3)

# Gradient Descent Parameters
alpha <- 0.1      # Learning rate
max_iter <- 100    # Maximum iterations
tolerance <- 1e-6  # Stopping criterion (tolerance)
x <- 0             # Initial guess (can be any value)

# Initialize history for tracking the process
history <- data.frame(iteration = integer(), x = numeric(), f_x = numeric())

# Gradient Descent Loop
for (i in 1:max_iter) {
  gradient <- grad_f(x)
  x_new <- x - alpha * gradient  # Update x with the gradient descent formula
  
  # Store history for visualization
  history <- rbind(history, data.frame(iteration = i, x = x_new, f_x = f(x_new)))
  
  # Check for convergence
  if (abs(x_new - x) < tolerance) {
    break
  }
  
  # Update x for next iteration
  x <- x_new
}

# Results
cat("Optimal x:", x, "\n")
cat("Minimum value of f(x):", f(x), "\n")
cat("Iterations to converge:", i, "\n")

# Plot Convergence
plot(history$iteration, history$f_x, type = "b", 
     xlab = "Iteration", ylab = "f(x)", 
     main = "Convergence of Gradient Descent")
abline(h = 0, col = "red", lty = 2)  # True minimum (f(x) = 0 at x = 3)
```

### Exp 14: Solve Convex Quadratic Optimization Problems

```r
# Install necessary libraries
install.packages("quadprog")
install.packages("quantmod")

# Load required libraries
library(quadprog)
library(quantmod)

set.seed(42)

# Number of assets
n_assets <- 4

# Synthetic returns (mean and covariance matrix)
mu <- c(0.08, 0.10, 0.12, 0.15)  # Expected returns
Sigma <- matrix(c(
  0.03, 0.01, 0.02, 0.01,
  0.01, 0.04, 0.02, 0.03,
  0.02, 0.02, 0.05, 0.04,
  0.01, 0.03, 0.04, 0.06
), nrow = n_assets, byrow = TRUE)  # Covariance matrix

# Parameters
target_return <- 0.10  # Minimum acceptable return

# Quadratic term (Dmat): 2 * covariance matrix (quadprog uses 1/2 * w' * D * w)
Dmat <- 2 * Sigma

# Linear term (dvec): Zero (since we only minimize variance)
dvec <- rep(0, n_assets)

# Constraints:
# 1. Budget constraint: sum(w_i) = 1
# 2. Return constraint: sum(w_i * mu_i) >= target_return
# 3. No short-selling: w_i >= 0
Amat <- t(rbind(
  rep(1, n_assets),  # Budget constraint
  mu                 # Return constraint
))

bvec <- c(
  1,                 # Budget = 1
  target_return,     # Target return
  rep(0, n_assets)   # No short-selling
)

# Equality constraint is just the first constraint (budget = 1)
meq <- 1 

# Solve the quadratic programming problem
solution <- solve.QP(Dmat, dvec, Amat, bvec, meq)

# Optimal weights
weights <- round(solution$solution, 4)
names(weights) <- paste("Asset", 1:n_assets)

# Portfolio statistics
portfolio_return <- sum(weights * mu)
portfolio_variance <- solution$value

# Output results
cat("Optimal Weights:\n")
print(weights)

cat("\nPortfolio Return:", portfolio_return, "\n")
cat("Portfolio Variance:", portfolio_variance, "\n")

# Barplot of optimal portfolio weights
barplot(weights,
        main = "Optimal Portfolio Weights",
        xlab = "Assets",
        ylab = "Weight",
        col = "steelblue")
```

### Exp 15: Stochastic Gradient Descent for Linear Regression (Minimizing MSE)

```r
# Generate synthetic data
set.seed(42)
X <- matrix(rnorm(100 * 2), ncol = 2)  # 100 samples, 2 features
true_weights <- c(3, -2)  # True coefficients
y <- X %*% true_weights + rnorm(100)  # Linear relation with noise

# SGD Parameters
learning_rate <- 0.01
epochs <- 100
weights <- c(0, 0)  # Initialize weights

# Plot setup to track convergence of weights
weight_history <- matrix(NA, nrow = epochs, ncol = 2)  # Store weights for each epoch

# SGD Loop
for (epoch in 1:epochs) {
  for (i in sample(nrow(X))) {  # Random sample (stochastic)
    # Compute gradient for the i-th sample
    prediction <- sum(X[i, ] * weights)
    error <- prediction - y[i]
    gradient <- error * X[i, ]
    
    # Update weights
    weights <- weights - learning_rate * gradient
  }
  
  # Store the weights at each epoch
  weight_history[epoch, ] <- weights
}

# Print the final estimated weights
cat("Estimated weights:", round(weights, 2), "\n")
cat("True weights:", true_weights, "\n")

# Plot the convergence of the weights
plot(1:epochs, weight_history[, 1], type = "l", col = "red", xlab = "Epoch", ylab = "Weight",
     main = "Convergence of Weights in SGD", ylim = range(weight_history))
lines(1:epochs, weight_history[, 2], col = "blue")
abline(h = true_weights[1], col = "red", lty = 2)  # True weight for first feature
abline(h = true_weights[2], col = "blue", lty = 2)  # True weight for second feature
legend("topright", legend = c("Estimated weight 1", "Estimated weight 2", "True weight 1", "True weight 2"),
       col = c("red", "blue", "red", "blue"), lty = c(1, 1, 2, 2))
```

### Exp 16: Stochastic Gradient Descent for Linear Regression (Binary Classification)

```r
# Generate synthetic data (binary classification)
set.seed(42)
X <- matrix(rnorm(100 * 2), ncol = 2)  # 100 samples, 2 features
true_weights <- c(1.5, -1)  # True coefficients

# Logistic transformation
logits <- X %*% true_weights
prob <- 1 / (1 + exp(-logits))  # Sigmoid function
y <- rbinom(100, 1, prob)  # Binary outcomes

# SGD Parameters
learning_rate <- 0.05
epochs <- 100
weights <- c(0, 0)  # Initialize weights

# SGD Loop
for (epoch in 1:epochs) {
  for (i in sample(nrow(X))) {  # Random sample (stochastic)
    # Compute gradient for logistic loss
    logit <- sum(X[i, ] * weights)
    pred_prob <- 1 / (1 + exp(-logit))  # Predicted probability
    gradient <- (pred_prob - y[i]) * X[i, ]  # Gradient of logistic loss
    
    # Update weights
    weights <- weights - learning_rate * gradient
  }

  # Optionally print estimated weights after each epoch (you can reduce the frequency if needed)
  if (epoch %% 10 == 0) {  # Print every 10 epochs
    cat("Epoch:", epoch, "Estimated weights:", round(weights, 2), "\n")
  }
}

# Final accuracy check
predicted_class <- ifelse(1 / (1 + exp(-X %*% weights)) > 0.5, 1, 0)  # Predicted class (0 or 1)
accuracy <- mean(predicted_class == y)  # Accuracy
cat("Final Accuracy:", round(accuracy, 2), "\n")
```
